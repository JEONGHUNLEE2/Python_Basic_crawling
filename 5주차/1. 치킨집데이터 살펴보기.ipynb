```python
import pandas as pd
```

```python
file = './data/2018년 전국 상가업소 업력현황.xlsx'
raw = pd.read_excel(file)
```

```python
raw.head()
```

```python
# 업종대분류 카테고리...
raw['업종대분류'].unique()
```

```python
# 업종중분류 카테고리
raw['업종중분류'].unique()
```

```python
### 추가 1. 업종대분류 '음식' 선택 -> 업종중분류 카테고리 살펴보기
###      2. 업종대분류, 업종중분류 인덱스 피벗데이블 
```

```python
# 업종대분류 " 음식 "
cond = raw['업종대분류'] == "음식"
raw[cond] # 업종대분류 " 음식 "
raw[cond]['업종중분류']
raw[cond]['업종중분류'].unique()
```

```python
# 업종중분류 '닭/오리요리' 인 데이터 선택

################### 조건선택하기 
cond = raw['업종중분류'] == '닭/오리요리'
df = raw[cond]
df
###
#raw[raw['업종중분류'] == '닭/오리요리']
```

```python
# 치킨집 수 합계 구하기 
#df['컬럼명1'] + df['컬럼명2'] + .......
df.describe()
```

```python
df['계'] = df.sum(axis = 1)
df
```

```python
# 실습
# 치킨집이 가장 많ㅇ느 곳?? --> 3곳
# 치킨집이 가장 적은곳

#sort_values()
df.sort_values(by = '계', ascending = False).head(3)
```
