```python
# 인구 데이터 서울특별시 종로구 (1111000000) --> "서울특별시 종로구 "
# 치킨 데이터 "서울특별시" "종로구" --> "서울특별시 종로구 "
```

```python
temp_list = []

for c in raw['행정기관'] :
    #print(c[:-13])
    temp = c.split(' (')[0] # 앞뒤로 나누어서 정리하는 방법
    #temp = c[:-13] 글자수로 정리하는 방법 
    print(temp)
    temp_list.append(temp)
    # 코드 정리 방법
    #### 길이수가 같으면 --> c[:-13]
    #### "()" 기준으로 데이터 split --> temp = c.split(' (')[0]
```

```python
temp_list
```

```python
raw['시군구정리'] = temp_list
raw.head()
```
