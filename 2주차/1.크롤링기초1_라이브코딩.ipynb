```python
search_list = ['부동산','파이썬','삼성','입시']
for word in search_list :
    print(word)
    url = f"https://search.daum.net/search?w=news&nil_search=btn&DA=NTB&enc=utf8&cluster=y&cluster_page=1&q={word}"
    print(url)
    browser.get(url)
#접속 방법 
#크롤링... 파일에 저장해줘..나중에 이렇게 쓰임
```
